```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read CSV
df = pd.read_csv("data/raw/mtsamples.csv", index_col=0)
```

```python
# Basic shape
print(df.shape)
print(df.columns.tolist())
# Should Expect: ['description', 'medical_specialty', 'sample_name', 'transcription', 'keywords']
```

```python
# Check for missing values (does according to excel)
print(df.isnull().sum())
```

```python
# Class distribution - Label Visibility
specialty_counts = df['medical_specialty'].value_counts()
print(specialty_counts)
print(f"\nTotal specialties: {df['medical_specialty'].nunique()}")
```

```python
# Report length distribution
df['token_count'] = df['transcription'].dropna().apply(lambda x: len(x.split()))
print(df['token_count'].describe())
```

```python
# How many exceed 512 tokens (BERT's limit)?
over_512 = (df['token_count'] > 512).sum()
print(f"Reports exceeding 512 tokens: {over_512} ({over_512/len(df)*100:.1f}%)")
```
